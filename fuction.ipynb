```python
def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
def div(a,b):
    print(a/b)
    
def cal(a,b,o):
    if o == "+":
        add(a,b)
    elif o == "-":
        sub(a,b)
    elif o == "*":
        mul(a,b)
    elif (o == "/") and (b!= 0):
        div(a,b)

    elif (o == "/") and (b== 0):
        print("The value of b can not be zero.")

    else:
        print("Invalid operator")
a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("Enter + or - or * or /")
cal(a,b,o)
```

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
    
def cal(a,b,o):
    if o == "+":
        print(add(a,b))
    elif o == "-":
        print(sub(a,b))
    elif o == "*":
        print(mul(a,b))
    elif (o == "/") and (b!= 0):
        print(div(a,b))

    elif (o == "/") and (b== 0):
        print("The value of b can not be zero.")

    else:
        print("Invalid operator")
        
a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("Enter + or - or * or /")
cal(a,b,o)
```

```python
def area(l,b):    
    a = l*b
    return a

def volume(l,b,h):
    v = area(l,b)*h
    return v

l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))
print(area(l,b))
print(volume(l,b,h))
```
